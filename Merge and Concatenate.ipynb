# Pandas Merge vs Concatenate

```python
import pandas as pd
set1 = pd.DataFrame({"A": [1,2,3,4,5],"B": [21,33,22,55,66]})

set2 = pd.DataFrame({"A":[1,2,3,4,9],"Z":[45,66,77,88,0]})

pd.merge(set1,set2,on="A")
```

# Merge supports four types of joins 

 * 1 . Left

```python
pd.merge(set1,set2,how="left")
```

* 2 . Right



```python
pd.merge(set1,set2,how="right")
```

* 3 . Outer

```python
pd.merge(set1,set2,how="outer")
```

* 4 . Inner

```python
pd.merge(set1,set2,how="inner")
```

```python
pd.merge(set1,set2,how="outer",indicator=True)
```

# Merge Using Indexes

```python
 pd.merge(set1,set2,how="outer",left_index=True,right_index=True)
```

# Handling Duplicate Column Names with Suffixes

```python
 pd.merge(set1,set2,how="outer",left_index=True,right_index=True,suffixes=("Number","Roll"))
```

# Concatenate

```python
set1 = pd.DataFrame({"A": [1,2,3,4,5],"B": [21,33,22,55,66]})

set2 = pd.DataFrame({"i":[1,2,3,4,9],"Z":[45,66,77,88,0]})

pd.concat([set1,set2],axis=1)
```

```python
set1 = pd.DataFrame({"A": [1,2,3,4,5],"B": [21,33,22,55,66]})

set2 = pd.DataFrame({"i":[1,2,3,66,77],"Z":[33,2,3,4,1]})

pd.concat([set1,set2],axis=1,join="inner")
```
